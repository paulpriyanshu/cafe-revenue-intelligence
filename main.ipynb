```python
import numpy as np
df["Total Spent"] = pd.to_numeric(df["Total Spent"],errors="coerce")
df.replace(['UNKNOWN', 'ERROR'], np.nan, inplace=True)
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors="coerce")
df['Quantity']       = pd.to_numeric(df['Quantity'],       errors='coerce')
# Case 1: Total Spent missing
mask_T = df['Total Spent'].isna() & df['Quantity'].notna() & df['Price Per Unit'].notna()
df.loc[mask_T, 'Total Spent'] = df.loc[mask_T, 'Quantity'] * df.loc[mask_T, 'Price Per Unit']

# mask_T = df['calculated_total'].isna() & df['Quantity'].notna() & df['Price Per Unit'].notna()
# df.loc[mask_T, 'calculated_total'] = df.loc[mask_T, 'Quantity'] * df.loc[mask_T, 'Price Per Unit']

# Case 2: Quantity missing
mask_Q = df['Quantity'].isna() & df['Total Spent'].notna() & df['Price Per Unit'].notna()
df.loc[mask_Q, 'Quantity'] = df.loc[mask_Q, 'Total Spent'] / df.loc[mask_Q, 'Price Per Unit']

# Case 3: Price Per Unit missing
mask_P = df['Price Per Unit'].isna() & df['Total Spent'].notna() & df['Quantity'].notna()
df.loc[mask_P, 'Price Per Unit'] = df.loc[mask_P, 'Total Spent'] / df.loc[mask_P, 'Quantity']


# df[df["Total Spent"]>10.0]
df[1000:1020]
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df = df.sort_values('Transaction Date').reset_index(drop=True)  # ← KEY STEP
df = df.dropna(subset=['Item', 'Total Spent', 'Quantity'])
df['calculated_total'] = df['Quantity'] * df['Price Per Unit']
df['price_mismatch']   = ~np.isclose(df['Total Spent'], df['calculated_total'], rtol=0.01)
df[1000:1020]
```

## 1. Practical Business Metrics
In this section, we calculate metrics that are essential for daily cafe management, including **Average Order Value (AOV)** and **Pareto (80/20) Analysis**.

```python
# Calculate Average Order Value (AOV) trends
df['Month'] = df['Transaction Date'].dt.to_period('M')
monthly_aov = df.groupby('Month')['Total Spent'].mean()
print('Monthly Average Order Value (AOV):\n', monthly_aov)

# Overall AOV
print('\nOverall average spend per item row:', df['Total Spent'].mean())
```

```python
import matplotlib.pyplot as plt

# Pareto (80/20) Analysis
item_revenue = df.groupby('Item')['Total Spent'].sum().sort_values(ascending=False)
item_cum_percent = 100 * item_revenue.cumsum() / item_revenue.sum()

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(item_revenue.index, item_revenue, color='skyblue')
ax1.set_ylabel('Total Revenue ($)', color='blue')

ax2 = ax1.twinx()
ax2.plot(item_revenue.index, item_cum_percent, color='red', marker='o', ms=5)
ax2.axhline(80, color='grey', linestyle='--')
ax2.set_ylabel('Cumulative Percentage (%)', color='red')

plt.title('Pareto Analysis of Item Revenue')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

top_80_items = item_cum_percent[item_cum_percent <= 85].index.tolist()
print(f'Items contributing to ~80% of revenue: {top_80_items}')
```

## 5. Multi-Horizon Sales Forecasting
Iterative prediction of revenue using the trained model.

```python
def predict_multi_horizon(model, daily_data, horizon_days):
    """Predicts revenue for a given number of days into the future."""
    temp_df = daily_data.copy()
    if 'Transaction Date' in temp_df.columns:
        temp_df = temp_df.set_index('Transaction Date')
    
    predictions = []
    for i in range(horizon_days):
        # Use the next date after the current sequence
        current_date = temp_df.index[-1] + pd.Timedelta(days=1)
        
        # Features must exactly match the training set: 
        # [lag1, lag2, lag3, lag7, day_of_week, month, rolling_mean_7, rolling_std_7, diff1]
        next_features = pd.DataFrame({
            'lag1': [temp_df['Revenue'].iloc[-1]],
            'lag2': [temp_df['Revenue'].iloc[-2]],
            'lag3': [temp_df['Revenue'].iloc[-3]],
            'lag7': [temp_df['Revenue'].iloc[-7]],
            'day_of_week': [current_date.dayofweek],
            'month': [current_date.month],
            'rolling_mean_7': [temp_df['Revenue'].tail(7).mean()],
            'rolling_std_7': [temp_df['Revenue'].tail(7).std() or 0],
            'diff1': [temp_df['Revenue'].iloc[-1] - temp_df['Revenue'].iloc[-2]]
        }, index=[current_date])
        
        # Match the model's expected type (float)
        pred = model.predict(next_features.astype(float))[0]
        predictions.append(max(0, pred))
        
        # Append prediction to dataframe for subsequent lags
        next_features['Revenue'] = pred
        temp_df = pd.concat([temp_df, next_features])
        
    return predictions

if 'best_model' in locals() and 'daily_df' in locals():
    # Perform projections
    next_day = predict_multi_horizon(best_model, daily_df, 1)
    next_week = predict_multi_horizon(best_model, daily_df, 7)
    next_month = predict_multi_horizon(best_model, daily_df, 30)

    print(f'--- REVENUE FORECASTS ---\n')
    print(f'Next Day Projection:   ${next_day[0]:,.2f}')
    print(f'Next Week (Cumulative): ${sum(next_week):,.2f}')
    print(f'Next Month (Cumulative): ${sum(next_month):,.2f}')
else:
    print('Error: "best_model" or "daily_df" not found. Check modeling cells.')
```

```python
daily_df['lag1'] = daily_df['Revenue'].shift(1)
daily_df['lag2'] = daily_df['Revenue'].shift(2)
daily_df['lag3'] = daily_df['Revenue'].shift(3)
daily_df['lag7'] = daily_df['Revenue'].shift(7)
daily_df['day_of_week'] = daily_df['Transaction Date'].dt.weekday
daily_df['month'] = daily_df['Transaction Date'].dt.month

print(daily_df['lag1'])
```

```python
X = daily_df[['lag1', 'lag2', 'lag3','lag7','day_of_week', 'month']]
y = daily_df['Revenue']
```

```python
# Add more features for better prediction
daily_df['rolling_mean_7'] = daily_df['Revenue'].rolling(window=7).mean()
daily_df['rolling_std_7'] = daily_df['Revenue'].rolling(window=7).std()
daily_df['diff1'] = daily_df['Revenue'].diff(1)

# Fill NaN values from rolling and diff operations
daily_df = daily_df.bfill()

# Update feature set
X = daily_df[['lag1', 'lag2', 'lag3', 'lag7', 'day_of_week', 'month', 'rolling_mean_7', 'rolling_std_7', 'diff1']]
y = daily_df['Revenue']

# Re-split the data
split = int(len(daily_df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=4)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, preds)
print("R² score:", r2)
```

```python
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV

# Time series cross-validation
tscv = TimeSeriesSplit(n_splits=3)

# Hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6],
    'subsample': [0.8, 0.9, 1.0]
}

# Grid search with time series CV
grid_search = GridSearchCV(
    XGBRegressor(random_state=42), 
    param_grid, 
    cv=tscv, 
    scoring='r2', 
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

# Best model
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)
print("Best CV score:", grid_search.best_score_)

# Evaluate on test set
best_preds = best_model.predict(X_test)
best_r2 = r2_score(y_test, best_preds)
print("Test R² with best model:", best_r2)
```
